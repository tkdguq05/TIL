```python
pd.date_range("2018-4-1", "2018-4-30")
```

```python
pd.date_range(start="2018-4-1", periods=30)
```

```python
pd.date_range("2018-4-1", "2018-4-30", freq="B") #B weekday
```

```python
pd.date_range("2018-1-1", "2018-12-31", freq="W") #W Sunday
```

```python
pd.date_range("2018-1-1", "2018-12-31", freq="W-MON")
```

```python
pd.date_range("2018-4-1", "2018-12-31", freq="MS") #MS first day of the month
```

```python
pd.date_range("2018-4-1", "2018-12-31", freq="M") # last day
```

```python
np.random.seed(0)
ts = pd.Series(np.random.randn(4), index=pd.date_range("2018-1-1", periods=4, freq="M"))
ts
```

```python
ts.shift(1)
```

```python
ts.shift(1, freq="M")
```

```python
ts.shift(1, freq="W")
```

```python
ts = pd.Series(np.random.randn(100), index=pd.date_range("2018-1-1", periods=100, freq="D"))
ts.tail(20)
```

```python
ts.resample('W').mean()
```

```python
ts.resample('M').first()
```

```python
ts = pd.Series(np.random.randn(60), index=pd.date_range("2018-1-1", periods=60, freq="T"))
ts.head(20)
```

```python
ts.resample('10T', closed="right").sum()
```

```python
ts.resample('5T').ohlc()
```

```python
ts.resample('30s').ffill().head(20)
```

```python
ts.resample('30s').bfill().head(20)
```
